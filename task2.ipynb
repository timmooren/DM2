```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data/training_set_VU_DM.csv', parse_dates=['date_time'])
```

```python
# show percentage of missing values and sort. ignore rows with no missing values
missing = df.isnull().sum() / len(df)
missing = missing[missing > 0]
missing.sort_values(inplace=True, ascending=False)
# export to latex
missing.to_latex('tex/missing.tex')
missing
```

```python
# plot the missing values as a barchart
plt.figure(figsize=(12, 6))
plt.title('Missing values')
plt.xlabel('Variables')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
sns.barplot(x=missing.index, y=missing.values * 100)
# show a dotted line at 100%
plt.axhline(y=100, color='black', linestyle='--')
plt.savefig('plots/missing.png', bbox_inches='tight')
```

```python
# show correlation matrix as heatmap
corr = df.corr()
plt.figure(figsize=(32, 32))
sns.heatmap(corr, annot=True, fmt='.2f', square=True, cmap='coolwarm')
plt.savefig('plots/corr.png')
plt.show()
```
